```python
import pandas as pd
import numpy as np
```

# collisions.csv

```python
data_folder = '../CSV-2018/'

collisions = pd.read_csv(data_folder+'collisions2018.csv', error_bad_lines=False, low_memory=False, dtype={'case_id': str})
```

```python
collisions.columns
```

```python
collisions.columns = ['case_id', 'col_date', 'col_severity', 'col_time',
       'county_city', 'hit_run', 'jurisdiction', 'lighting',
       'loc_type', 'officer_id', 'pcf_violation_code',
       'pcf_violation_category', 'subsection', 'population',
       'pcf_type', 'process_date', 'ramp_int',
       'road_con_1', 'road_con_2', 'road_surf', 'tow_away',
       'col_type', 'weather_1', 'weather_2']
```

```python
collisions[collisions.case_id.duplicated(keep=False)].sort_values('case_id')
```

```python
collisions
```

```python
collisions.isnull().any()
```

### road_en

```python
road_en = collisions[['road_con_1']].append(collisions[['road_con_2']].rename(columns={'road_con_2': 'road_con_1'}), ignore_index=True)
road_en = pd.DataFrame(set(road_en.road_con_1), columns=['road_con']).dropna().sort_values('road_con').reset_index(drop=True)
road_1 = collisions.merge(road_en.rename(columns={'road_con': 'road_con_1'}), on=['road_con_1'], how='left')
road_1 = road_1[['case_id', 'road_con_1']].dropna().sort_values('case_id').reset_index(drop=True)
road_2 = collisions.merge(road_en.rename(columns={'road_con': 'road_con_2'}), on=['road_con_2'], how='left')
road_2 = road_2[['case_id', 'road_con_2']].dropna().sort_values('case_id').reset_index(drop=True)
road_en = pd.concat([road_1.rename(columns={'road_con_1': 'road_con'}), road_2.rename(columns={'road_con_2': 'road_con'})], ignore_index=True).sort_values('case_id').reset_index(drop=True)
road_en
```

```python
road_en.to_csv(r'../data/road_en.csv', index=False)
```

### weather_en

```python
weather_en = collisions[['weather_1']].append(collisions[['weather_2']].rename(columns={'weather_2': 'weather_1'}), ignore_index=True)
weather_en = pd.DataFrame(set(weather_en.weather_1), columns=['weather']).dropna().sort_values('weather').reset_index(drop=True)
weather_1 = collisions.merge(weather_en.rename(columns={'weather': 'weather_1'}), on=['weather_1'], how='left')
weather_1 = weather_1[['case_id', 'weather_1']].dropna().sort_values('case_id').reset_index(drop=True)
weather_2 = collisions.merge(weather_en.rename(columns={'weather': 'weather_2'}), on=['weather_2'], how='left')
weather_2 = weather_2[['case_id', 'weather_2']].dropna().sort_values('case_id').reset_index(drop=True)
weather_en = pd.concat([weather_1.rename(columns={'weather_1': 'weather'}), weather_2.rename(columns={'weather_2': 'weather'})], ignore_index=True).sort_values('case_id').reset_index(drop=True)
weather_en
```

```python
weather_en.to_csv(r'../data/weather_en.csv', index=False)
```

### location

```python
location = collisions[['population', 'county_city','loc_type', 'ramp_int']]
location = location.drop_duplicates().reset_index(drop=True)
location = location.reset_index()
location.columns = ['loc_num', 'population', 'county_city','loc_type', 'ramp_int']
location
```

```python
location.to_csv(r'../data/location.csv', index=False)
```

### pcf

```python
pcf = collisions[['pcf_violation_code', 'pcf_violation_category', 'subsection', 'pcf_type']]
pcf = pcf.drop_duplicates().reset_index(drop=True)
pcf = pcf.reset_index()
pcf = pcf.rename(columns={'index': 'pcf_num'})
pcf
```

```python
pcf.to_csv(r'../data/pcf.csv', index=False)
```

### case

```python
collide_in = collisions.merge(location, on=['population', 'county_city','loc_type', 'ramp_int'], how='left')
collide_in = collide_in[['case_id', 'loc_num']]
collide_in
```

```python
cause = collisions.merge(pcf, on=['pcf_violation_code', 'pcf_violation_category', 'subsection', 'pcf_type'], how='left')
cause = cause[['case_id', 'pcf_num']]
cause
```

```python
case = collisions[['case_id', 'col_date', 'col_severity', 'col_time', 'hit_run',\
                   'jurisdiction', 'officer_id', 'process_date', 'tow_away', 'col_type','lighting', 'road_surf']]
case = case.merge(collide_in, on='case_id').merge(cause, on='case_id')
case['col_time'] = case['col_date'] + ' ' + case['col_time']
case = case.drop(['col_date'], axis=1)
case
```

```python
case.to_csv(r'../data/case.csv', index=False)
```

# parties.csv

```python
parties = pd.read_csv(data_folder+'parties2018.csv', error_bad_lines=False, low_memory=False, dtype={'case_id': str})
```

```python
parties.columns = ['at_fault', 'case_id', 'phone', 'fin_resp',
       'haz_mat', 'party_id', 'move_pre','other_fac_1', 'other_fac_2', 
       'age','drug_phy', 'party_num', 'safety_equip_1',
       'safety_equip_2', 'sex', 'sobriety', 'party_type',
       'school_bus_rel', 've_type', 've_make', 've_year']
```

### other_fac_en

```python
other_fac_en = parties[['other_fac_1']].append(parties[['other_fac_2']].rename(columns={'other_fac_2': 'other_fac_1'}), ignore_index=True)
other_fac_en = pd.DataFrame(set(other_fac_en.other_fac_1), columns=['other_fac']).dropna().sort_values('other_fac').reset_index(drop=True)
other_fac_1 = parties.merge(other_fac_en.rename(columns={'other_fac': 'other_fac_1'}), on=['other_fac_1'], how='left')
other_fac_1 = other_fac_1[['party_id', 'other_fac_1']].dropna().sort_values('party_id').reset_index(drop=True)
other_fac_2 = parties.merge(other_fac_en.rename(columns={'other_fac': 'other_fac_2'}), on=['other_fac_2'], how='left')
other_fac_2 = other_fac_2[['party_id', 'other_fac_2']].dropna().sort_values('party_id').reset_index(drop=True)
other_fac_en = pd.concat([other_fac_1.rename(columns={'other_fac_1': 'other_fac'}), other_fac_2.rename(columns={'other_fac_2': 'other_fac'})], ignore_index=True).sort_values('party_id').reset_index(drop=True)
other_fac_en
```

```python
other_fac_en.to_csv(r'../data/other_fac_en.csv', index=False)
```

### vehicle

```python
vehicle = parties[['ve_type', 've_make', 've_year','school_bus_rel']].drop_duplicates().reset_index(drop=True)\
                .reset_index().rename(columns={'index': 've_num'})
vehicle
```

```python
vehicle.to_csv(r'../data/vehicle.csv', index=False)
```

### party

```python
take = parties.merge(vehicle, on=['ve_type', 've_make', 've_year','school_bus_rel'], how='left')
take = take[['party_id', 've_num']]
take
```

```python
party = parties[['party_id', 'case_id', 'at_fault', 'phone', 'fin_resp',
       'haz_mat', 'move_pre', 'age','drug_phy', 'party_num', 'sex', 'sobriety', 'party_type']]
party = party.merge(take, on='party_id')
party = party.drop_duplicates().reset_index(drop=True)
party
```

```python
party.to_csv(r'../data/party_involve.csv', index=False)
```

### safety_p

```python
safety_p = parties[['safety_equip_1']].append(parties[['safety_equip_2']].rename(columns={'safety_equip_2': 'safety_equip_1'}), ignore_index=True)
safety_p = pd.DataFrame(set(safety_p.safety_equip_1), columns=['safety_equip']).dropna().sort_values('safety_equip').reset_index(drop=True)
safety_p_1 = parties.merge(safety_p.rename(columns={'safety_equip': 'safety_equip_1'}), on=['safety_equip_1'], how='left')
safety_p_1 = safety_p_1[['party_id', 'safety_equip_1']].dropna().sort_values('party_id').reset_index(drop=True)
safety_p_2 = parties.merge(safety_p.rename(columns={'safety_equip': 'safety_equip_2'}), on=['safety_equip_2'], how='left')
safety_p_2 = safety_p_2[['party_id', 'safety_equip_2']].dropna().sort_values('party_id').reset_index(drop=True)
safety_p = pd.concat([safety_p_1.rename(columns={'safety_equip_1': 'safety_equip'}), safety_p_2.rename(columns={'safety_equip_2': 'safety_equip'})], ignore_index=True).sort_values('party_id').reset_index(drop=True)
safety_p
```

```python
safety_p.to_csv(r'../data/safety_equip_en.csv', index=False)
```

# victims.csv

```python
victims = pd.read_csv(data_folder+'victims2018.csv', error_bad_lines=False, low_memory=False, dtype={'case_id': str})
```

```python
victims.head()
```

```python
victims.isnull().any()
```

```python
victims.columns = ['case_id', 'vic_id', 'party_num', 'vic_age','deg_injury', 'ejected', 
                      'vic_role','safety_equip_1', 'safety_equip_2','vic_seat', 'vic_sex']
```

### victim

```python
associate = victims[['case_id', 'party_num','vic_id']].merge(parties, on=['case_id', 'party_num'])
associate = associate[['vic_id','party_id']].drop_duplicates().reset_index(drop=True)
associate
```

```python
victim = victims[['vic_id', 'vic_age', 'deg_injury', 'ejected', 'vic_role', 'vic_seat', 'vic_sex']]
victim = victim.merge(associate, on=['vic_id'])
victim
```

```python
victim.to_csv(r'../data/associate_victim.csv', index=False)
```

### safety_v

```python
safety_v = victims[['safety_equip_1']].append(victims[['safety_equip_2']].rename(columns={'safety_equip_2': 'safety_equip_1'}), ignore_index=True)
safety_v = pd.DataFrame(set(safety_v.safety_equip_1), columns=['safety_equip']).dropna().sort_values('safety_equip').reset_index(drop=True)
safety_v_1 = victims.merge(safety_v.rename(columns={'safety_equip': 'safety_equip_1'}), on=['safety_equip_1'], how='left')
safety_v_1 = safety_v_1[['vic_id', 'safety_equip_1']].dropna().sort_values('vic_id').reset_index(drop=True)
safety_v_2 = victims.merge(safety_v.rename(columns={'safety_equip': 'safety_equip_2'}), on=['safety_equip_2'], how='left')
safety_v_2 = safety_v_2[['vic_id', 'safety_equip_2']].dropna().sort_values('vic_id').reset_index(drop=True)
safety_v = pd.concat([safety_v_1.rename(columns={'safety_equip_1': 'safety_equip'}), safety_v_2.rename(columns={'safety_equip_2': 'safety_equip'})], ignore_index=True).sort_values('vic_id').reset_index(drop=True)
safety_v
```

```python
safety_v.to_csv(r'../data/have_vs.csv', index=False)
```
